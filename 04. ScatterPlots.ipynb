# Gráficos de Dispersión

En este JNB se revisará el tema de Gráficos de Dispersión.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
insurance_filepath = "https://raw.githubusercontent.com/GTD2101-ACD-GA-AD22/seaborn-ricardo-qm/main/data/insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data
```

# Gráficas de dispersión

Este tipo de gráfico muestra la relación que pudieran tener parejas de datos (columnas).

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

#Diagrama de dispersión codificado por color

Permite mostrar la relación entre más de dos variables (de hecho entre tres variables). 

Vamos a mostrar de que manera el fumar afecta la relación entre el bmi (indice de masa corporal) y el charge (cargo de la poliza). Mostrando los datos clasificados por color dependiendo de si se fuma o no.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

Se pueden crear diagramas de dispersión para el caso de variables categóricas (aquéllas que no son numéricas)

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```
