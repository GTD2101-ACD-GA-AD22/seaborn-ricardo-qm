# Gráficos de Distribución

En este JNB se revisará el tema de Gráficos de Distribución.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
iris_filepath = "https://raw.githubusercontent.com/GTD2101-ACD-GA-AD22/seaborn-ricardo-qm/main/data/iris.csv"

# Read the file into a variable iris_data
iris_data = pd.read_csv(iris_filepath, index_col="Id")

# Print the first 5 rows of the data
iris_data.head()
```

# Histogramas

Los histogramas permiten graficar la frecuencia de los datos mediante barras de frecuencia.

```python
# Histogram 
sns.histplot(iris_data['Petal Length (cm)'])
```

```python
# Histogram 
sns.histplot(iris_data['Petal Width (cm)'])
```

```python
# Histogram 
sns.histplot(iris_data['Sepal Length (cm)'])
```

```python
# Histogram 
sns.histplot(iris_data['Sepal Width (cm)'])
```

# Diagramas de Densitty - Histogramas suavizados

Este tipo de diagrama -llamado Kernel Density Estimate (KDE) es una forma suavizada de los histogramas.

```python
# KDE plot 
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

```python
# KDE plot 
sns.kdeplot(data=iris_data['Petal Width (cm)'], shade=True)
```

```python
# KDE plot 
sns.kdeplot(data=iris_data['Sepal Length (cm)'], shade=True)
```

```python
# KDE plot 
sns.kdeplot(data=iris_data['Sepal Width (cm)'], shade=True)
```

# Diagramas KDE - 2D

Un tipo de Diagrama que permite mostrar histogramas suavizados para dos o más valores. Bi-dimensionales.

```python
# 2D KDE plot
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
```

```python
# 2D KDE plot
sns.jointplot(x=iris_data['Petal Width (cm)'], y=iris_data['Sepal Length (cm)'], kind="kde")
```

```python
# 2D KDE plot
sns.jointplot(y=iris_data['Petal Length (cm)'], x=iris_data['Sepal Width (cm)'], kind="kde")
```

```python
# 2D KDE plot
sns.jointplot(y=iris_data['Petal Width (cm)'], x=iris_data['Sepal Length (cm)'], kind="kde")
```
